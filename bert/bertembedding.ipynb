```python
from transformers import BertTokenizer,BertModel
```

```python
import torch
from torch import nn
```

```python
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
test_sent = 'this is a test sentence'
```

```python
input = tokenizer(test_sent,return_tensors='pt')
```

```python
input
```

```python
input_ids = input['input_ids']
token_type_ids = input['token_type_ids']
```

输出的是一个三维张量（Tensor），形状通常为 [batch_size, sequence_length, hidden_size],输出以 grad_fn=<NativeLayerNormBackward0> 结尾，这表明在词嵌入层的最后一步进行的是层归一化（Layer Normalization）操作，并且 PyTorch 正在记录计算图以便进行反向传播求导。

```python
model.embeddings(input_ids,token_type_ids)
```

这一行代码在手动生成位置索引（Position IDs）。

input_ids.shape 通常返回一个元组 (batch_size, sequence_length)。

input_ids.shape[1] 提取的是序列长度（sequence_length）。

torch.arange(n) 是 PyTorch 中的一个函数，用于生成一个从 0 开始到 n-1 的一维整数张量。

```python
pos_ids = torch.arange(input_ids.shape[1])
```

```python
pos_ids
```

```python
token_embed = model.embeddings.word_embeddings(input_ids)
```

```python
token_embed
```

```python
token_embed.shape
```

```python
seg_embed = model.embeddings.token_type_embeddings(token_type_ids)
```

```python
seg_embed
```

```python
seg_embed.shape
```

```python
pos_embed = model.embeddings.position_embeddings(pos_ids)
```

```python
pos_embed
```

```python
pos_embed.shape
```

pos_embed.unsqueeze(0) 的作用是在张量的第 0 个维度（最外层）增加一个大小为 1 的“批次”（Batch）维度，以便与其他特征张量进行形状对齐和相加。

```python
input_embed = token_embed + seg_embed + pos_embed.unsqueeze(0)
```

```python
input_embed
```

```python
embed = model.embeddings.LayerNorm(input_embed)
```

```python
embed = model.embeddings.dropout(embed)
```

经过了上述三个embedding相加后，进行layernorm归一再dropout，得到的结果就等于model.embeddings(input_ids,token_type_ids)的输出
