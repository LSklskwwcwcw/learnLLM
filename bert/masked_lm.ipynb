```python
import torch
from torch import nn
from transformers.models.bert import BertModel,BertTokenizer,BertForMaskedLM
```

```python
model_type = 'bert-base-uncased'
```

```python
tokenizer = BertTokenizer.from_pretrained(model_type)
bert = BertModel.from_pretrained(model_type)
mlm = BertForMaskedLM.from_pretrained(model_type,output_hidden_states=True)
```

```python
text = ("After Abraham Lincoln won the November 1860 presidential "
        "election on an anti-slavery platform, an initial seven "
        "slave states declared their secession from the country "
        "to form the Confederacy. War broke out in April 1861 "
        "when secessionist forces attacked Fort Sumter in South "
        "Carolina, just over a month after Lincoln's "
        "inauguration.")
```

```python
text
```

```python
inputs = tokenizer(text,return_tensors='pt')
```

```python
inputs
```

```python
inputs['input_ids'].shape
```

inputs['input_ids'][0]：提取单条数据的 ID
在上一行代码中，你打印了 inputs['input_ids'].shape，结果是 [1, 62]。这表示当前的数据是一个批次（Batch），批次大小为 1，包含了 62 个词元的 ID。

加上 [0] 相当于进行切片降维，把外层代表 Batch 的那个维度去掉，仅仅提取出这 62 个数字组成的一维数组（Tensor）。

tokenizer.convert_ids_to_tokens(...)：数字转词元
这是 Hugging Face 分词器自带的一个极其常用的方法。

模型是不认识英文字母的，它只认识数字（比如 101, 2023 等）。这个方法相当于去查字典（Vocabulary），把这 62 个数字 ID 一一映射回对应的英文词块（Token）列表。

在这个阶段，输出的其实是一个 Python 列表（List），里面全是单个的字符串，比如 ['[CLS]', 'after', 'abraham', ...]。

' '.join(...)：拼接成完整字符串
这是 Python 原生的字符串操作方法。

它把上一步得到的包含 62 个字符串的列表，用空格（' '）作为分隔符，首尾相连拼成一个完整的长字符串，最终打印在屏幕上。

```python
' '.join(tokenizer.convert_ids_to_tokens(inputs['input_ids'][0]))
```

# masking

```python
inputs['labels'] = inputs['input_ids'].detach().clone()
```

```python
inputs['labels']
```

```python
mask = torch.rand(inputs['input_ids'].shape) < 0.15
```

```python
mask
```

```python
sum(mask[0])
```

```python
10/62
```

这里取出了第一句话的掩码数组 mask[0]，并使用 sum() 计算里面有多少个 True。

结果算出来有 10 个词元被挑中了。

接着做了一个简单的数学除法 10 / 62。结果是 16.1%，这非常接近我们设定的 15% 的目标概率，证明随机生成是符合预期的。
这行代码就是为了解决这个问题：

条件 1：(torch.rand(...) < 0.15)
和前面一样，先保证 15% 的基础随机选中率。

条件 2：(inputs['input_ids'] != 101)
在 BERT (基于 bert-base-uncased) 的词表中，数字 101 代表 [CLS]。这个条件保证了当前词元不是 [CLS] 时，结果才为 True。

条件 3：(inputs['input_ids'] != 102)
同理，数字 102 代表 [SEP]。这个条件保证了当前词元不是 [SEP] 时，结果才为 True。

* (乘法符号)：
在 PyTorch 中，当两个（或多个）布尔类型的张量进行相乘时，它代表的是逻辑与（AND）操作。

整段逻辑翻译成白话就是：

必须同时满足三个条件，一个词元才会被标记为 True（被遮掩）：

恰好命中了 15% 的随机概率。 并且

它的 ID 不是 101（不是 [CLS]）。 并且

它的 ID 不是 102（不是 [SEP]）。

```python
mask_arr = (torch.rand(inputs['input_ids'].shape)< 0.15) \
      * (inputs['input_ids'] != 101) \
      * (inputs['input_ids'] != 102)
```

```python
mask_arr
```

```python
sum(mask_arr[0])
```

mask_arr[0]：取出第一句话的布尔掩码数组（里面有 10 个 True，其余是 False）。

.nonzero()：这是一个非常实用的函数，它会找出数组中所有非零元素（在布尔数组中，True 就是 1，非零）的索引位置。

torch.flatten() 和 .tolist()：将提取出来的张量索引展平为一维，并转换成标准的 Python 列表。

结果：列表里的这 10 个数字，准确地指出了上一条代码中随机选中的 10 个词汇在这个句子里的具体位置（比如第 11 个词、第 14 个词等）。

```python
selection = torch.flatten(mask_arr[0].nonzero()).tolist()
```

```python
selection
```

```python
tokenizer.special_tokens_map
```

这两行代码我们在上一个问题中已经接触过。它首先查阅字典，确认了 [MASK] 这个特殊符号在模型里的数字代号是 103。

随后，它定位到之前随机选出的 15% 的位置（也就是 selection 列表里的索引），把 input_ids 里原本的真实单词 ID 强行替换成了 103。

注意打印出的 inputs 字典：你会发现在 input_ids 这个张量里，确实多出了很多 103。但是，在 labels 这个张量里，对应位置依然保留着最原始的单词 ID。

```python
tokenizer.vocab['[MASK]']
```

```python
inputs['input_ids'][0,selection] = 103
```
