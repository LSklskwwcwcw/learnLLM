```python
s1 = 'albums sold 123122342132342424 copies'
s2 = 'technically perfect, melodically correct'
s3 = 'featuring a previously unheard track'
s4 = 'bestselling music artist'
s5 = 's1 d1 o1 and o2'
s6 = 'asbofwheohwbeif'
```

```python
from transformers import BertTokenizer
```

```python
model_name = 'bert-base-uncased'
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
tokenizer.vocab
```

```python
tokenizer.convert_ids_to_tokens
```

tokenizer不会轻易的处理文本为unk

```python
inputs = tokenizer(s1)
inputs
```

```python
tokenizer.decode(inputs['input_ids'])
```

```python
tokenizer.convert_ids_to_tokens(inputs['input_ids'])
```

## 基于词汇表,tokenize,encode,decode一体
- tokenize:word => tokens
- encode:token => id
- decode: id => token => word
 - encode之后也不是终点,decode要尽可能很好的将id还原,尽可能与输入的word对齐

```python
inputs = tokenizer(s6)
print(inputs)
print(tokenizer.convert_ids_to_tokens(inputs['input_ids']))
```
